```python
period_map = {
    'Unknown': 'Unknown',
    'Ancient': 'Ancient',
    'Prehistoric': 'Prehistoric',
    '6th century AD': 'Ancient',
    '7th century': 'Ancient',
    '10th century': 'Medieval',
    '11th century': 'Medieval',
    '12th century': 'Medieval',
    '14th century': 'Medieval',
    '16th century': 'Medieval',
    '1960s': 'Contemporary'
}
```

```python
def convert_to_period(x):
    try:
        # Try to cast to int (works for 1800, 1950, etc.)
        year = int(x)
        if year < 500:
            return 'Ancient'
        elif year < 1500:
            return 'Medieval'
        elif year < 1900:
            return 'Modern'
        else:
            return 'Contemporary'
    except:
        # If not a valid int, map manually or default to 'Unknown'
        return period_map.get(x)
```

```python
data['established']=data['established'].apply(convert_to_period)
```

```python
user_input = pd.DataFrame([{
    'zone': 'Eastern',
    'type': 'Entertainment',
    'established': 'Modern',
    'time to visit': 'Moderate',
    'entrance fee': 'Free',
    'airport nearby': 'No',
    'dslr allowed': 'No',
    'best time': 'Morning',
    'foot fall': 'Low'
}])
```

```python
zone_value = user_input.iloc[0]['zone']
filtered_data = data[data['zone'] == zone_value]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.neighbors import NearestNeighbors
```

```python
cols=data.drop(columns=['name']).columns
```

```python
preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore',drop='first',sparse_output=False), cols),
],remainder='drop')
```

```python
X_all = data.drop(columns=['name'])
preprocessor.fit(X_all)
```

```python
if filtered_data.empty or filtered_data.shape[0] < 1:
    print("❌ No recommendations found for the selected zone/state.")
else:
    X_filtered = filtered_data.drop(columns=['name'])
    X_filtered_processed = preprocessor.transform(X_filtered)

    knn = NearestNeighbors(
    n_neighbors=min(3, len(filtered_data)),
    metric='cosine',
    algorithm='brute',   # 'brute' is preferred for cosine
    n_jobs=-1
)
    knn.fit(X_filtered_processed)

    user_processed = preprocessor.transform(user_input)
    distances, indices = knn.kneighbors(user_processed)

    recommended_places = filtered_data.iloc[indices[0]]['name'].values

```

```python
if len(recommended_places) > 0:
    print("✅ Recommendations in", user_input.iloc[0]['zone'], ":")
    for name in recommended_places:
        print("-", name)
else:
    print("❌ No recommendations found in your selected region.")
```
